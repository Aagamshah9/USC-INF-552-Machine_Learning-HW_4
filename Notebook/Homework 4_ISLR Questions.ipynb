(a) According to this setting (x11=x12=x1 and x21=x22=x2), the ridge regression problem seeks to minimize

(y1−β^1x1−β^2x1)2+(y2−β^1x2−β^2x2)2+λ(β^21+β^22).

(b) By taking the derivatives of the above expression with respect to β^1 and β^2 and setting them equal to 0, we obtain  respectively

β^1(x21+x22+λ)+β^2(x21+x22)=y1x1+y2x2

and

β^1(x21+x22)+β^2(x21+x22+λ)=y1x1+y2x2.

By substracting the two expressions above we get β^1=β^2.

(c) According to this setting (x11=x12=x1 and x21=x22=x2), the lasso optimization problem seeks to minimize

(y1−β^1x1−β^2x1)2+(y2−β^1x2−β^2x2)2+λ(|β^1|+|β^2|).

(d) We will use the alternate form of the lasso optimization problem

(y1−β^1x1−β^2x1)2+(y2−β^1x2−β^2x2)2 subject to |β^1|+|β^2|≤s.

Geometrically the lasso constraint take the form of a diamond centered at the origin of the plane (β^1,β^2) which intersects the axes at a distance s from the origin. By using the setting of this problem (x11=x12=x1, x21=x22=x2, x1+x2=0 and y1+y2=0), we have to minimize the expression

2[y1−(β^1+β^2)x1]2≥0.

This optimization problem has a simple solution : β^1+β^2=y1/x1. Geometrically, this is a line parallel to the edge of the diamond of the constraints. Now, solutions to the lasso optimization problem are contours of the function [y1−(β^1+β^2)x1]2 that intersects the diamond of the constraints. So, the entire edge β^1+β^2=s (as is the edge β^1+β^2=−s) is a potential solution to the lasso optimization problem. Thus, the lasso optimization problem has a whole set of solutions instead of a unique one :

{(β^1,β^2):β^1+β^2=s with β^1,β^2≥0 and β^1+β^2=−s with β^1,β^2≤0}.
